```python
#step 3:Handling missing data or values.
#scenari-1:
#what we do value miss in the numerical value.
#we have the calculate mean or average. we can place the avg value to empty value
#missing in numerical column we have to calculate mean or average of that colum
#replace the avg or mean value in the empty place
#scenario-2:
#missing value in numerical column we have to replace in empty value with
#high repeated value.
#scenario-3:
#having missing value in usage date set.removing empty value rows from the data
import numpy as np # np is an instance numpy
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
#simpleImputer is class from sklearn . which is used to new value placed in the assimg value
#simpleImpute and sklearn are package
#simpleImpute is a class which was import the sklearn package
```

```python
X
```

```python
#step 4:handling the categorical data
# converting the categorical data in to numerical data.
from sklearn.preprocessing import OneHotEncoder
#OneHotEncoder is a class which was important from sklearn package
#OneHotEncoder is used to convert text to binary code (0,1)
from sklearn.compose import ColumnTransformer
#column transformer is a class which was important from sklearn package
#column transformer is used to apply change or transformer (updating the valuses present in the data is transformer)
#column transformer is a class which was used to apply the change of numeric data in the place of categorical
#srinu-0,0,1
#encoder is a method which convert text to numeric
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=ct.fit_transform(X)
```

```python
X
```

```python
#encoder the output column.(OneHotEncoder is used input columns. used for output encoder)
#OneHotEncoder is apply for input column
#labelencoder is only for output column

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
#step 5:split the data into training and testing data
#train_test_slipt is class which was imported from sklearn.model_selection
#It was uswd for sliping the data into training and test

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```
